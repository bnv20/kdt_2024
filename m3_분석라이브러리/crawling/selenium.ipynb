```python
# Implicit Waits(암묵적 대기) : 찾으려는 element가 로드될 때까지 지정한 시간만큼 대기할 수 있도록 설정
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get(url='https://www.google.com/')
try:
    element = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CLASS_NAME , 'gLFyf'))
    )
    # 요소가 존재하면 텍스트 출력 (이 경우에는 입력창이므로 text가 없을 수 있음)
    print("Element found:", element.get_attribute('title'))
except Exception as e:
    # 예외 발생 시 메시지 출력
    print("An error occurred:", e)
finally:
    # WebDriver 종료
    driver.quit()
```

```python
# 1. find_element(By.ID, "id_name"): id 속성을 사용하여 접근
try:
    element_by_id = driver.find_element(By.ID, "APjFqb")
    print("Element found by ID:", element_by_id.get_attribute('name')) # 찾은 요소의 alt 속성 값을 가져온다.
except:
    print("Element not found by ID")
```

```python
# 2. find_element(By.CLASS_NAME, "class_name"): 클래스를 사용하여 접근

try:
    element_by_class_name = driver.find_element(By.CLASS_NAME, "gLFyf")
    print("Element found by CLASS_NAME:", element_by_class_name.get_attribute('title'))
except:
    print("Element not found by CLASS_NAME")
```

```python
# 3. find_element(By.NAME, "name"): name 속성을 사용하여 접근
try:
    element_by_name = driver.find_element(By.NAME, "q")
    print("Element found by NAME:", element_by_name.get_attribute('title'))
except:
    print("Element not found by NAME")
```

```python
# 4. find_element(By.XPATH, "xpath"): xpath 속성을 사용하여 접근
try:
    element_by_xpath = driver.find_element(By.XPATH, '//*[@id="APjFqb"]')
    print("Element found by XPATH:", element_by_xpath.get_attribute('title'))
except:
    print("Element not found by XPATH")
```

1. getElementById()
 - 파라미터로 찾으려는 id를 전달하여, 해당 element를 찾을 수 있습니다. id는 유일한 값이므로, 하나의 element만 리턴합니다.
2. getElementsByClassName()
 - 클래스 이름으로 element를 찾아서,이 클래스 이름을 가지는 모든 element 목록을 리턴합니다.이 함수의 이름을 자세히 보면 getElementsByClassName으로 Element's'가 복수 형태인 것은 이 함수가 목록을 리턴하기 때문입니다.
3. getElementByTagName()
 - 위 코드는 'div' 태그를 가지는 모든 element 목록을 찾아줍니다.이 함수 역시, 다수의 element를 리턴하기 때문에 함수명에 복수형의 'elements'가 포함되어 있습니다.
4. querySelector()
 - DOM에서 원하는 element를 찾기 위해서 querySelector() 를 사용할 수도 있는데, 이 함수는 파라미터로 입력받은 CSS선택자를 사용해서, element를 찾아줍니다.
 - querySelector() 함수는, 파라미터로 입력받은 CSS 선택자로 찾은 여러개의 element 중 첫번째 element를 리턴합니다.
 - 태그 이름으로 element를 찾을 때는 태그명을 문자열로 넘겨줍니다.위 예제는 div태그를 가지는 element 중 첫번째 element를 리턴합니다.
5. querySelectorAll()
 - 사용법은 querySelector() 와 같습니다. 다만, querySelectorAll()은 이름에서 알수 있듯이, CSS선택자(파라미터로 넘겨준)로 찾은 모든 element 목록을 리턴합니다.

```python
import requests
from pandas import DataFrame
from bs4 import BeautifulSoup
import re
from datetime import datetime
import os

date = str(datetime.now())
date = date[:date.rfind(':')].replace(' ', '_')
date = date.replace(':','시') + '분'

query = input('검색 키워드를 입력하세요 : ')
news_num = int(input('총 필요한 뉴스기사 수를 입력해주세요(숫자만 입력) : '))
query = query.replace(' ', '+') # 검색 키워드 공백 처리

news_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}'

req = requests.get(news_url.format(query)) # format() 메서드를 사용하여 {} 부분을 query 변수로 대체
soup = BeautifulSoup(req.text, 'html.parser')


news_dict = {}
idx = 0
cur_page = 1

print()
print('크롤링 중...')

while idx < news_num:
### 네이버 뉴스 웹페이지 구성이 바뀌어 태그명, class 속성 값 등을 수정###

    # class 속성 값이 list_news인 ul 태그를 찾아서 table 변수에 저장
    table = soup.find('ul',{'class' : 'list_news'})
    # table 변수에서 id 속성 값이 sp_nws로 시작하는 모든 li 태그를 찾아서 li_list 변수에 저장
    li_list = table.find_all('li', {'id': re.compile('sp_nws.*')})
    # li_list 변수에서 각각의 li 태그 안에서 class 속성 값이 news_area인 div 태그를 찾아서 area_list 변수에 저장
    area_list = [li.find('div', {'class' : 'news_area'}) for li in li_list]
    # area_list 변수에서 각각의 div 태그 안에서 class 속성 값이 news_tit인 a 태그를 찾아서 a_list 변수에 저장
    a_list = [area.find('a', {'class' : 'news_tit'}) for area in area_list]
    # a_list 변수에서 가져올 뉴스 기사의 개수를 news_num과 현재까지 가져온 뉴스 기사의 개수인 idx 중 작은 값으로 설정
    for n in a_list[:min(len(a_list), news_num-idx)]:  # news_num-idx는 news_num과 idx 의 차이
        news_dict[idx] = {'title' : n.get('title'),
                          'url' : n.get('href') }
        idx += 1
    # cur_page += 1: 현재 검색 중인 페이지 번호를 1 증가
    cur_page += 1

pages = soup.find('div', {'class' : 'sc_page_inner'})
next_page_url = [p for p in pages.find_all('a') if p.text == str(cur_page)][0].get('href')

req = requests.get('https://search.naver.com/search.naver' + next_page_url)
soup = BeautifulSoup(req.text, 'html.parser')

print('크롤링 완료')

print('데이터프레임 변환')
news_df = DataFrame(news_dict).T

news_df
```

    - datetime.now(): 현재 시간을 가져옵니다.
    - date.rfind(':'): 문자열 date에서 마지막으로 등장하는 :의 인덱스를 가져옵니다.
    - date[:date.rfind(':')]: 문자열 date에서 마지막 :을 기준으로 앞부분만 가져옵니다.

    - date.replace(':', '시') + '분': 문자열 date에서 :을 시로 대체하고, 뒤에 분을 추가합니다.
- 사용자로부터 검색 키워드와 총 필요한 뉴스 기사의 개수를 입력받습니다.
    - input(): 사용자로부터 값을 입력받습니다.
    - query.replace(' ', '+'): 검색 키워드에서 공백을 +로 대체합니다.
    - requests.get(): 지정된 URL에서 HTTP GET 요청을 전달합니다.
    - soup = BeautifulSoup(req.text, 'html.parser'): HTTP GET 요청의 응답으로 받은 HTML 코드를 BeautifulSoup 객체로 파싱합니다.

- 검색 결과 페이지의 번호를 나타낼 변수 cur_page를 생성
- while 루프를 실행하여 검색 결과를 가져옵니다.
    - while 루프의 조건식 idx < news_num은 가져온 뉴스 기사 개수가 news_num보다 작을 때 반복합니다.
    - table = soup.find('ul', {'class': 'list_news'}): 웹페이지에서 뉴스 기사 목록을 포함하는 ul 태그를 가져옵니다.
    - li_list = table.find_all('li', {'id': re.compile('sp_nws.*')}): 뉴스 기사 목록에서 각각의 뉴스 기사를 포함하는 li 태그를 가져옵니다.
    - area_list = [li.find('div', {'class': 'news_area'}) for li in li_list]: 뉴스 기사 목록에서 각각의 뉴스 기사를 포함하는 div 태그를 가져옵니다.
    - a_list = [area.find('a', {'class': 'news_tit'}) for area in area_list]: 뉴스 기사 목록에서 각각의 뉴스 기사를 나타내는 a 태그를 가져옵니다.
    - for n in a_list[:min(len(a_list), news_num - idx)]:: 현재 페이지에서 가져올 뉴스 기사 개수가 news_num - idx보다 크면 news_num - idx개만 가져옵니다.
    - news_dict[idx] = {'title': n.get('title'), 'url': n.get('href')}: 딕셔너리 news_dict에 뉴스 기사의 제목과 URL을 저장합니다.
    - cur_page += 1: 검색 결과 페이지의 번호를 1 증가시킵니다.    
    - pages = soup.find('div', {'class': 'sc_page_inner'}): 다음 검색 결과 페이지의 URL을 가져오기 위해 현재 검색 결과 페이지에서 페이지 네이게이션을 포함하는 div 태그를 가져옵니다.
    - next_page_url = [p for p in pages.find_all('a') if p.text == str(cur_page)][0].get('href'): 다음 검색 결과 페이지의 URL을 가져옵니다. pages.find_all('a') 메서드를 사용하여 모든 페이지 네이게이션 링크를 찾은 후, str(cur_page)와 일치하는 링크를 찾습니다.
    - req = requests.get('https://search.naver.com/search.naver' + next_page_url): 다음 검색 결과 페이지의 URL을 사용하여 HTTP GET 요청을 보냅니다.
    - soup = BeautifulSoup(req.text, 'html.parser'): HTTP 응답으로부터 HTML 코드를 추출하여 BeautifulSoup 객체를 생성합니다.
- 검색 결과를 딕셔너리 news_dict에 저장한 후, news_dict를 데이터프레임으로 변환합니다.
    - DataFrame(news_dict).T는 딕셔너리 news_dict를 데이터프레임으로 변환합니다. .T는 데이터프레임을 전치시키는 역할을 합니다.
    - to_excel(file_name)은 데이터프레임을 엑셀 파일로 저장합니다.

    -  os.getcwd()는 현재 작업 디렉토리를 반환합니다.
    - print('엑셀 저장 완료 | 경로 : {}\\{}'.format(dir_path, file_name))는 엑셀 파일이 저장된 경로를 출력합니다.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
from datetime import datetime

date = str(datetime.now())
date = date[:date.rfind(':')].replace(' ', '_')
date = date.replace(':', '시') + '분'

query = input('검색 키워드를 입력하세요 : ')
news_num = int(input('총 필요한 뉴스기사 수를 입력해주세요(숫자만 입력) : '))
query = query.replace(' ', '+') # 검색 키워드 공백 처리

news_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}'

req = requests.get(news_url.format(query))
soup = BeautifulSoup(req.text, 'html.parser')

news_dict = {}
idx = 0
cur_page = 1

print()
print('크롤링 중...')

while idx < news_num:

    table = soup.find('ul', {'class': 'list_news'})
    li_list = table.find_all('li', {'id': re.compile('sp_nws.*')})
    area_list = [li.find('div', {'class': 'news_area'}) for li in li_list]
    a_list = [area.find('a', {'class': 'news_tit'}) for area in area_list]

    for n in a_list[:min(len(a_list), news_num-idx)]:
        news_dict[idx] = {'title': n.get('title'),
                          'url': n.get('href')}
        idx += 1

    cur_page += 1

    # 다음 페이지 URL 추출
    pages = soup.find('div', {'class': 'sc_page_inner'})
    next_page_url = [p for p in pages.find_all('a') if p.text == str(cur_page)][0].get('href')

    # 다음 페이지 HTML 코드 가져오기
    req = requests.get('https://search.naver.com/search.naver' + next_page_url)
    soup = BeautifulSoup(req.text, 'html.parser')

print('크롤링 완료')

# 데이터프레임 변환
news_df = pd.DataFrame(news_dict).T

# Excel 파일 저장
filename = '네이버뉴스_{}_{}.xlsx'.format(query, date)
sheetname = 'Sheet1'
news_df.to_excel(filename, sheet_name=sheetname, index=False)

print(f'파일 저장 완료: {filename}')
```
