## 통계분석과 기계학습

통계분석
- 통계분석은 데이터로부터 모델을 만들어 현상을 설명하고 이해하는 데 중점
- 단순하고 해석 가능한 형태. 이 모델들은 명확한 가정(예: 정규성, 독립성, 등분산성)에 기반하고, 이러한 가정들은 모델의 유효성을 판단하는 데 중요
- 통계분석의 결과는 데이터에 대한 해석과 추론에 중점

기계학습
-  데이터로부터 패턴을 학습하여 미래 데이터에 대한 예측을 하는 데 중점
-  복잡하고, 통계적 가정보다는 예측 성능에 초점을 둔다. 이러한 모델들은 때로 해석하기 어려울 수 있으며, "블랙 박스"로 간주
- 기계학습의 결과는 주로 예측의 정확도와 모델의 성능에 중점. 해석 가능성보다는 예측 성능이 더 중요한 역할

머신러닝과 인공지능에는 여러 가지 통계적 개념과 기법들이 필요

1. 기초 통계 개념
- 기술 통계 (Descriptive Statistics): 데이터의 기본적인 특성을 요약하고 설명합니다. 평균, 중앙값, 표준편차, 분산, 범위, 사분위수 등이 포함.
- 추론 통계 (Inferential Statistics): 표본 데이터를 이용해 모집단에 대한 결론을 도출. 가설 검정, 신뢰 구간, p-value 등이 포함.
2. 확률 이론 (Probability Theory)
- 확률 분포 (Probability Distributions): 데이터가 어떤 패턴을 따르는지 설명합니다. 정규분포, 이항분포, 포아송분포 등이 포함.
- 확률 변수 (Random Variables): 결과가 확률에 따라 결정되는 변수. 이산 확률 변수와 연속 확률 변수로 나눌 수 있다.
- 기대값과 분산 (Expected Value and Variance): 확률 변수의 평균과 변동성을 측정.
3. 가설 검정 (Hypothesis Testing)
- 귀무가설과 대립가설 (Null and Alternative Hypothesis): 검정의 기본 가정. 귀무가설은 보통 '효과 없음'을 가정.
- t-검정 (t-test): 두 그룹의 평균을 비교하는 데 사용.
- 카이제곱 검정 (Chi-Square Test): 범주형 데이터의 분포를 비교.
- ANOVA (Analysis of Variance): 세 개 이상의 그룹 간의 평균 차이를 비교.
4. 회귀 분석 (Regression Analysis)
- 단순 선형 회귀 (Simple Linear Regression): 하나의 독립 변수와 종속 변수 간의 관계를 모델링.
- 다중 선형 회귀 (Multiple Linear Regression): 여러 독립 변수와 종속 변수 간의 관계를 모델링.
- 로지스틱 회귀 (Logistic Regression): 종속 변수가 범주형일 때 사용.
5. 확률 모델링 (Probabilistic Modeling)
- 베이지안 통계 (Bayesian Statistics): 사전 지식과 데이터를 결합하여 결론을 도출.
- 마르코프 체인 (Markov Chain): 상태 전환 확률을 모델링하는 데 사용.
6. 고급 통계 기법
- 주성분 분석 (PCA, Principal Component Analysis): 고차원의 데이터를 저차원으로 축소하여 중요한 패턴을 찾는다.
- 클러스터링 (Clustering): 데이터 포인트를 유사한 그룹으로 나눈다. K-평균, 계층적 클러스터링 등이 포함.
- 서포트 벡터 머신 (SVM, Support Vector Machine): 데이터 분류에 사용되는 강력한 기법.
7. 시간 시계열 분석 (Time Series Analysis)
- 추세 분석 (Trend Analysis): 데이터의 장기적인 증가 또는 감소 경향을 파악.
- 계절성 분석 (Seasonality Analysis): 데이터의 주기적인 패턴을 분석.
- ARIMA 모델 (AutoRegressive Integrated Moving Average): 시간 시계열 데이터를 분석하고 예측하는 데 사용.

## 기초 통계

평균과 표준편차

데이터의 중심 경향성과 분산을 파악하는 데 유용. 예를 들어, 학생들의 시험 점수 데이터를 분석할 때 평균과 표준편차를 통해 전체 성적 수준과 변동성을 이해할 수 있다.

```python
import numpy as np

# 예제 데이터
data = np.array([85, 90, 78, 92, 88, 76, 95, 89, 77, 91])

# 평균 및 표준편차 계산
mean = np.mean(data)
std_dev = np.std(data)

print(f"Mean: {mean}, Standard Deviation: {std_dev}")
```

중간값과 사분위수

데이터의 중심 경향과 분포를 이해하는 데 도움을 준다. 예를 들어, 가계 소득 데이터를 분석할 때 중간값과 사분위수를 통해 소득 분포를 파악할 수 있다.

```python
# 중간값 및 사분위수 계산
median = np.median(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

print(f"Median: {median}, Q1: {q1}, Q3: {q3}")
```

평균값 > 중앙값
- 데이터는 오른쪽으로 치우친(오른쪽 꼬리가 긴) 분포를 가진다. 이를 '오른쪽으로 치우친 분포' 또는 '양의 왜곡 분포'라고 한다.
- 특징: 데이터의 높은 값(오른쪽 꼬리)에 이상치나 극단적인 높은 값들이 존재한다.
- 예시: 적은 수의 매우 높은 소득을 가진 사람들이 있는 소득 데이터. 이러한 극단적으로 높은 값들은 평균을 중앙값보다 더 높게 밀어올린다.

중앙값 > 평균값
- 데이터는 일반적으로 왼쪽으로 치우친(왼쪽 꼬리가 긴) 분포를 가진다. 이러한 분포를 '왼쪽으로 치우친 분포' 또는 '음의 왜곡 분포'라고 한다.
- 특징: 데이터의 낮은 값(왼쪽 꼬리)에 이상치나 극단적인 낮은 값들이 존재할 수 있다.
- 예시: 적은 수의 매우 낮은 소득을 가진 사람들이 있는 소득 데이터. 이러한 극단적으로 낮은 값들은 평균을 중앙값보다 더 낮게 끌어내린다.

중요성

평균과 중앙값 사이의 차이는 데이터의 비대칭성을 나타내며 이 차이가 클수록 데이터 분포의 왜곡이 더 크다고 볼 수 있다. 데이터 분석에서는 이러한 왜곡을 인지하고 적절한 해석과 분석 방법을 선택하는 것이 중요. 예를 들어, 이상치나 극단적인 값이 데이터의 해석에 큰 영향을 미칠 수 있으므로, 이를 고려하여 분석을 진행.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 생성
data_right_skewed = np.random.beta(a=2, b=10, size=1000)  # 오른쪽으로 긴 꼬리
data_left_skewed = np.random.beta(a=10, b=2, size=1000)   # 왼쪽으로 긴 꼬리

# DataFrame 생성
df_right_skewed = pd.DataFrame(data_right_skewed, columns=['val'])
df_left_skewed = pd.DataFrame(data_left_skewed, columns=['val'])

# 히스토그램 그리기
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# 오른쪽으로 긴 꼬리 분포
axs[0].hist(df_right_skewed['val'], bins=30, color='blue', alpha=0.7)
axs[0].axvline(np.mean(df_right_skewed['val']), color='red', linestyle='dashed', linewidth=1)
axs[0].axvline(np.median(df_right_skewed['val']), color='green', linestyle='dashed', linewidth=1)
axs[0].legend({'Mean':np.mean(df_right_skewed['val']), 'Median':np.median(df_right_skewed['val'])})
axs[0].set_title('Right-Skewed Distribution')

# 왼쪽으로 긴 꼬리 분포
axs[1].hist(df_left_skewed['val'], bins=30, color='orange', alpha=0.7)
axs[1].axvline(np.mean(df_left_skewed['val']), color='red', linestyle='dashed', linewidth=1)
axs[1].axvline(np.median(df_left_skewed['val']), color='green', linestyle='dashed', linewidth=1)
axs[1].legend({'Mean':np.mean(df_left_skewed['val']), 'Median':np.median(df_left_skewed['val'])})
axs[1].set_title('Left-Skewed Distribution')

plt.tight_layout()
plt.show()


```

의미와 중요성

- 이상치의 영향: 평균값은 이상치에 의해 크게 왜곡될 수 있다. 예를 들어, 소득 데이터에서 몇몇 매우 높은 값(억만장자의 소득)이 평균을 크게 올릴 수 있지만, 중앙값은 이러한 극단적인 값의 영향을 덜 받는다.
- 데이터 분포의 대표성: 중앙값은 데이터의 분포가 비대칭일 때 데이터의 "중심"을 더 잘 나타내는 경향이 있다. 반면, 평균값은 데이터가 대칭적인 분포를 가질 때 중심 경향을 잘 나타낸다.
- 데이터 해석: 데이터 분석에서는 종종 평균값과 중앙값을 함께 사용하여 데이터의 분포를 더 잘 이해한다. 두 척도 사이에 큰 차이가 있다면, 이는 데이터에 이상치나 극단적인 값이 있다는 신호일 수 있다.
따라서 평균값과 중앙값을 선택할 때는 데이터의 특성과 분석의 목적을 고려해야 하며 이상치가 존재하거나 데이터가 비대칭적인 분포를 가질 경우, 중앙값이 더 적절할 수 있다. 반면, 데이터가 정규 분포를 이루는 경우, 평균값이 데이터를 대표하는 데 적합할 수 있다.

산점도
두 변수 간의 관계를 시각화하는 데 유용. 예를 들어, 광고비와 매출 데이터를 분석할 때 산점도를 통해 광고비와 매출 간의 관계를 파악할 수 있다.

```python
# 예제 데이터
x = np.arange(100)
x = x
y = x + np.random.normal(0, 10, len(x))

# 산점도 그리기
plt.scatter(x, y)
plt.title('Scatter Plot of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

추세선이 있는 산점도
두 변수 간의 관계를 시각화하고, 추세선을 통해 데이터의 경향성을 파악하는 데 유용. 예를 들어, 마케팅 캠페인 기간과 매출 데이터를 분석할 때 추세선이 있는 산점도를 통해 캠페인 기간이 매출에 미치는 영향을 파악할 수 있다.

```python
# 추세선이 있는 산점도 그리기
plt.scatter(x, y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red')
plt.title('Scatter Plot with Trend Line')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

상관 계수
두 변수 간의 선형 관계를 측정하는 데 사용. 예를 들어, 온도와 아이스크림 판매량 데이터를 분석할 때 상관 계수를 통해 두 변수 간의 관계 강도를 파악할 수 있다.

```python
# 상관 계수 계산
correlation = np.corrcoef(x, y)
print(f"Correlation coefficient: {correlation[0, 1]}")
```

히트맵
변수 간의 상관 관계를 시각화하는 데 유용. 예를 들어, 여러 주식의 일일 수익률 데이터를 분석할 때 히트맵을 통해 각 주식 간의 상관 관계를 시각화할 수 있다.

```python
import numpy as np
import pandas as pd

# 가상의 실험 데이터 생성
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(0, 1, 100),
    'C': np.random.normal(0, 1, 100)
})

# 상관관계 행렬 계산
correlation_matrix = data.corr()
print(correlation_matrix)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 히트맵 그리기
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix for Chemical Compounds A, B, C')
plt.show()
```

- 변수 간의 상관관계: 각 화합물 A, B, C 간의 상관관계를 시각적으로 확인할 수 있습니다.
- 상관계수의 크기: 상관계수가 높을수록 강한 관계를 나타내며, 이는 실험 설계와 해석에 중요한 정보를 제공합니다.

## 확률

분산(Variance)과 표준편차(Standard Deviation)
- 통계학에서 데이터의 분포를 측정하는 데 사용되는 두 가지 중요한 개념.
- 이들은 데이터가 평균값으로부터 얼마나 퍼져 있는지(즉, 얼마나 분산되어 있는지)를 나타낸다.
- 분산과 표준편차는 데이터의 변동성을 이해하는 데 중요한 도구이며 이러한 측정값을 통해 데이터 집합의 일관성과 예측 가능성을 파악할 수 있다.

분산 (Variance)
- 정의: 데이터 포인트들이 평균에서 얼마나 멀리 떨어져 있는지를 측정하는 값.수학적으로 분산은 각 데이터 포인트와 평균값의 차이를 제곱한 후, 이를 모든 데이터 포인트의 개수로 나누어 평균을 낸 값.
- 의미: 분산이 크다는 것은 데이터 포인트들이 평균에서 멀리 떨어져 있다는 것을 의미하며, 이는 데이터가 넓게 퍼져 있다는 것을 나타낸다. 분산이 작다는 것은 데이터 포인트들이 평균값에 가깝게 모여 있다는 것을 의미하며 이는 데이터가 좁게 집중되어 있다는 것을 나타낸다.

표준편차 (Standard Deviation)
- 정의: 표준편차는 분산의 제곱근으로, 데이터 포인트들이 평균에서 얼마나 멀리 떨어져 있는지를 나타내는 또 다른 방법.
분산의 단위는 원래 데이터의 단위의 제곱이 되지만, 표준편차는 원래 데이터와 동일한 단위를 갖는다.
- 의미: 표준편차가 크다는 것은 데이터 포인트들이 평균에서 멀리 떨어져 있다는 것을 의미하며, 데이터가 넓게 분포되어 있다는 것을 나타낸다. 표준편차가 작다는 것은 데이터 포인트들이 평균값에 가까이 모여 있다는 것을 의미하며, 데이터가 좁게 집중되어 있다는 것을 나타낸다.

분산
- 불편분산 : 불편분산(Unbiased Variance)은 표본 데이터를 사용하여 모집단의 분산을 추정할 때 편향을 최소화하기 위해 사용되는 개념
 * 일반적으로 표본 분산은 모분산을 과소평가하는 경향이 있는데, 불편분산은 이러한 편향을 조정
- Numpy와 Pandas는 서로 다른 분산을 기본으로 설정하여 계산.
 * Numpy는 표본분산, Pandas는 불편분산
 * Pandas에서 표본분산을 계산하려면 var 메소드의 인수를 ddof=0이라고 설정, 불편분산은 ddof=1일 때 해당됨

[표본분산이 분산을 과소평가하는 이유]

- 분산은 '데이터와 평균값의 간의 거리' 이므로 모평균 간의 거리를 계산 : np.sum((fish_data - mu)**2)/N  
호수안에 물고기가 7마리 있다. 몸길이 {1,2,3,4,5,6,7}, 모평균 4,  
샘플링 3마리 {1,2,3} 표본평균 2  
이때 분산은 모평균을 모르기 때문에 표본평균과의 거리를 계산하며 분산을 과소평가 하게된다.
- 표본 분산 보다 조금 더 큰 값이 되도록 보정하기 위하여 N 대신 N-1로 나누어 불편분산을 구할 수 있다.

#### 정규분포

- 연속 확률분포 중 하나로, 데이터가 평균을 중심으로 좌우 대칭적인 형태를 띱니다.
- 중심극한정리에 의해 많은 자연 현상과 실험 결과가 정규분포를 따릅니다.






```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 데이터 생성
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# 시각화
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, stat="density", bins=30)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution (mean=0, std_dev=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

정규분포는 평균 주위에 데이터가 많이 모여 있으며, 멀어질수록 데이터의 빈도가 급격히 감소합니다. 이는 많은 자연 현상에서 평균이 가장 일반적이고, 극단적인 값은 드물다는 것을 의미합니다.

#### 이항분포
- 이항분포는 성공 또는 실패의 두 가지 결과만 있는 실험을 n번 반복했을 때, 성공 횟수의 분포를 나타냅니다. 성공 확률 p와 실험 횟수 n에 의해 결정됩니다.


```python
# 성공 확률이 0.5인 실험을 10번 반복한 이항분포를 시각화
from scipy.stats import binom

# 데이터 생성
n = 10
p = 0.5
x = np.arange(0, n+1)
prob = binom.pmf(x, n, p)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(x, prob, color='blue')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

이항분포는 주사위 던지기, 동전 던지기와 같은 이산적인 사건을 모델링하는 데 유용합니다. 성공 확률이 0.5인 경우, 성공 횟수가 5회 근처에 많이 분포함을 의미합니다.

#### 포아송분포
- 포아송분포는 일정 시간 또는 공간에서 발생하는 사건의 수를 모델링합니다. 사건의 평균 발생률 λ에 의해 결정됩니다.

```python
# 평균 발생률이 3인 포아송분포를 시각화
from scipy.stats import poisson

# 데이터 생성
lambda_ = 3
x = np.arange(0, 15)
prob = poisson.pmf(x, lambda_)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(x, prob, color='green')
plt.title('Poisson Distribution (λ=3)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

포아송분포는 일정한 시간 동안 발생하는 사건의 수를 예측하는 데 사용됩니다. 예를 들어, 한 시간 동안 발생하는 전화 수나 도착하는 고객 수 등을 모델링할 수 있습니다.

#### 기하분포
- 기하분포는 첫 성공까지의 실패 횟수를 모델링합니다. 성공 확률 p에 의해 결정됩니다.

```python
# 성공 확률이 0.3인 경우의 기하분포를 시각화합니다.
from scipy.stats import geom

# 데이터 생성
p = 0.3
x = np.arange(1, 15)
prob = geom.pmf(x, p)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(x, prob, color='purple')
plt.title('Geometric Distribution (p=0.3)')
plt.xlabel('Number of Trials Until First Success')
plt.ylabel('Probability')
plt.show()
```

기하분포는 첫 성공이 나올 때까지의 시도 횟수를 모델링합니다. 이는 고객에게 제품을 판매하기 위해 필요한 평균 시도 횟수를 예측하는 데 유용합니다.

#### 지수분포
- 지수분포는 사건 간의 시간 간격을 모델링합니다. 사건의 평균 발생률 λ에 의해 결정됩니다.

```python
# 평균 발생률이 1/2인 지수분포를 시각화합니다.
from scipy.stats import expon

# 데이터 생성
lambda_ = 0.5
data = np.random.exponential(1/lambda_, 1000)

# 시각화
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, stat="density", bins=30)
plt.title('Exponential Distribution (λ=0.5)')
plt.xlabel('Time Between Events')
plt.ylabel('Density')
plt.show()
```

지수분포는 두 사건 간의 시간 간격을 모델링합니다. 이는 서버의 고장 시간 간격이나 고객이 도착하는 시간 간격을 예측하는 데 유용합니다.

Q. 한 회사에서 임의로 선택한 50명의 직원들의 월급을 조사했더니 평균이 $3,500, 표준편차가 $600이었습니다. 이 회사의 전체 직원 월급의 95% 신뢰구간을 구하세요.

```python
import numpy as np
from scipy.stats import norm

# 표본 평균과 표준편차
sample_mean = 3500
sample_std = 600
n = 50

# 95% 신뢰구간 계산
z = norm.ppf(0.975)  # 95% 신뢰수준의 Z값
margin_of_error = z * (sample_std / np.sqrt(n))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(f"95% Confidence Interval: ${confidence_interval[0]:.2f} to ${confidence_interval[1]:.2f}")

# 인사이트
# 신뢰구간을 이용해 모집단 평균을 추정할 수 있습니다.
# 월급의 95% 신뢰구간은 $3,334.17에서 $3,665.83 사이입니다.
# 이는 회사 전체 직원의 월급 평균이 이 범위 내에 있을 가능성이 높음을 의미합니다.
```

인사이트

월급의 95% 신뢰구간이 $3,334.17에서 $3,665.83 사이로, 이 범위 내에서 전체 직원의 월급 평균이 있을 가능성이 높습니다. 이는 경영진이 직원들의 월급 수준을 평가하고 조정하는 데 유용한 정보를 제공합니다

## 가설 검정

Q. 한 제약회사는 새로 개발한 약물이 기존 약물보다 효과가 더 높은지 테스트하고자 합니다. 100명의 환자에게 새 약물을, 100명의 환자에게 기존 약물을 투여했더니, 새 약물 그룹의 평균 효과가 75, 표준편차가 10, 기존 약물 그룹의 평균 효과가 70, 표준편차가 15였습니다. 두 그룹 간의 평균 효과 차이가 유의미한지 95% 신뢰수준에서 검정하세요.

```python
from scipy.stats import ttest_ind

# 데이터 설정
new_drug_effect = np.random.normal(75, 10, 100)
old_drug_effect = np.random.normal(70, 15, 100)

# 독립 표본 t-검정 수행
t_stat, p_value = ttest_ind(new_drug_effect, old_drug_effect)

print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")

# 인사이트
# 가설검정을 통해 두 그룹 간의 평균 차이를 분석할 수 있습니다.
# p-value가 0.05보다 작으면 두 그룹 간의 차이가 유의미하다고 볼 수 있습니다.
if p_value < 0.05:
    print("새 약물이 기존 약물보다 효과가 유의미하게 높습니다.")
else:
    print("새 약물과 기존 약물의 효과 차이는 유의미하지 않습니다.")
```

인사이트

p-value가 0.05보다 작으므로, 새 약물의 효과가 기존 약물보다 유의미하게 높다는 결론을 내릴 수 있습니다. 이는 제약회사가 새 약물을 상용화할 수 있는 근거를 제공하며, 의사결정에 중요한 역할을 합니다.

Q. 세 가지 학습 방법(A, B, C)을 사용한 학생들의 시험 점수가 다음과 같습니다:

A 방법: [85, 90, 78, 92, 88]<BR>
B 방법: [80, 85, 88, 90, 84]<BR>
C 방법: [78, 82, 84, 88, 90]<BR>
세 학습 방법 간의 평균 시험 점수 차이가 유의미한지 분석하세요.

```python
from scipy.stats import f_oneway

# 데이터 설정
group_a = np.array([85, 90, 78, 92, 88])
group_b = np.array([80, 85, 88, 90, 84])
group_c = np.array([78, 82, 84, 88, 90])

# 일원분산분석 (ANOVA) 수행
f_stat, p_value = f_oneway(group_a, group_b, group_c)

print(f"F-statistic: {f_stat:.2f}, P-value: {p_value:.4f}")

# 인사이트
# ANOVA를 통해 여러 그룹 간의 평균 차이를 분석할 수 있습니다.
# p-value가 0.05보다 작으면 그룹 간의 차이가 유의미하다고 볼 수 있습니다.
if p_value < 0.05:
    print("세 학습 방법 간의 평균 점수 차이는 유의미합니다.")
else:
    print("세 학습 방법 간의 평균 점수 차이는 유의미하지 않습니다.")
```

인사이트

p-value가 0.05보다 크므로, 세 학습 방법 간의 평균 점수 차이가 유의미하지 않다는 결론을 내릴 수 있습니다. 이는 학습 방법에 따라 성과 차이가 크지 않음을 의미하며, 학습 방법 선택 시 다른 요인들을 고려할 수 있습니다.

Q. 한 마케팅 회사는 두 가지 광고 캠페인(A와 B)이 고객 행동에 미치는 영향을 비교하고자 합니다. 각각의 광고를 본 후 제품을 구매한 고객 수는 다음과 같습니다:

광고 A: 구매 30명, 비구매 70명<BR>
광고 B: 구매 20명, 비구매 80명<BR>
두 광고 간에 유의미한 차이가 있는지 검정하세요.

```python
from scipy.stats import chi2_contingency

# 데이터 설정
observed = np.array([[30, 70], [20, 80]])

# 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(observed)

print(f"Chi2 statistic: {chi2:.2f}, P-value: {p_value:.4f}")

# 인사이트
# 카이제곱 검정을 통해 두 광고 캠페인 간의 차이를 분석할 수 있습니다.
# p-value가 0.05보다 작으면 두 광고 간의 차이가 유의미하다고 볼 수 있습니다.
if p_value < 0.05:
    print("두 광고 캠페인 간에 유의미한 차이가 있습니다.")
else:
    print("두 광고 캠페인 간에 유의미한 차이가 없습니다.")
```

인사이트

p-value가 0.05보다 크므로, 두 광고 캠페인 간의 유의미한 차이가 없다는 결론을 내릴 수 있습니다. 이는 광고 A와 광고 B가 고객 행동에 미치는 영향이 비슷하다는 것을 의미합니다. 마케팅 전략을 수정하거나 다른 요인을 고려할 필요가 있습니다.

## 회귀 분석

Q. 한 회사의 마케팅 예산과 매출 데이터가 다음과 같습니다:

마케팅 예산(단위: 천 달러): [10, 20, 30, 40, 50]<BR>
매출(단위: 천 달러): [15, 25, 35, 45, 55]<BR>
마케팅 예산이 매출에 미치는 영향을 분석하기 위해 선형회귀 분석을 수행하고, 마케팅 예산이 매출에 미치는 영향에 대한 인사이트를 도출하세요.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# 데이터 설정
marketing_budget = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
sales = np.array([15, 25, 35, 45, 55])

# 선형회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(marketing_budget, sales)

# 회귀 계수와 절편 출력
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope: {slope:.2f}, Intercept: {intercept:.2f}")

# 인사이트
# 선형회귀를 통해 마케팅 예산이 매출에 미치는 영향을 분석할 수 있습니다.
# 기울기 (slope)가 1.0, 절편 (intercept)이 5.0으로, 마케팅 예산이 1천 달러 증가할 때마다 매출이 1천 달러 증가함을 나타냅니다.
print(f"마케팅 예산이 1천 달러 증가할 때마다 매출이 {slope:.2f}천 달러 증가합니다.")
```

인사이트

마케팅 예산이 매출에 긍정적인 영향을 미치며, 마케팅 예산이 1천 달러 증가할 때마다 매출이 1천 달러 증가한다는 것을 나타냅니다. 이는 회사가 마케팅 예산을 늘림으로써 매출 증대를 기대할 수 있음을 시사합니다.
