```python
# JSON
import pandas as pd
import json

# 데이터프레임 생성
data = {
    "name": ["Wes", "Scott", "Katie"],
    "age": [30, 31, 32],
    "pet": [None, "Zeus", "Sixes"]
}
df = pd.DataFrame(data)

# 데이터프레임을 JSON 파일로 저장
json_file_path = 'data.json'
df.to_json(json_file_path, orient='records', lines=True)

# JSON 파일을 다시 데이터프레임으로 불러오기
df_loaded = pd.read_json(json_file_path, orient='records', lines=True)

# 결과 출력
print(df,'\n')
print(df_loaded)
```
